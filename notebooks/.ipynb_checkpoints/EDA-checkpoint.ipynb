```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Purchased', data=df)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
